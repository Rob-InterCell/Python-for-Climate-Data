```python
print(df["TAVG"].describe())
```

```python
df["TAVG_C"] = (df["TAVG"] - 32) * 5/9
```

```python
print(df[["TAVG", "TAVG_C"]].head())
```

```python
print(df["TAVG"].describe())
```

```python
# Drop rows with missing temperature values
df = df.dropna(subset=["temperature"])
```

```python
# Drop rows with missing temperature values
df = df.dropna(subset=["TAVG"])
```

```python
print(df["TAVG_C"].describe())
```

```python
# Drop rows with missing temperature values
df = df.dropna(subset=["TAVG_C"])
```

```python
# Set DATE column as index and ensure it's parsed as datetime
df = df.set_index("DATE")

# Monthly average temperature (in Celsius)
monthly_avg = df["TAVG_C"].resample("M").mean()

# Add month column
df["month"] = df.index.month

# Map months to meteorological seasons
season_map = {
    12: "DJF", 1: "DJF", 2: "DJF",   # Winter
    3: "MAM", 4: "MAM", 5: "MAM",   # Spring
    6: "JJA", 7: "JJA", 8: "JJA",   # Summer
    9: "SON", 10: "SON", 11: "SON"  # Autumn
}
df["season"] = df["month"].map(season_map)

# Seasonal descriptive statistics for average temperature
seasonal_stats = df.groupby("season")["TAVG_C"].describe()
print(seasonal_stats)
```

```python
monthly_avg = df["TAVG_C"].resample("ME").mean()
```

```python
import matplotlib.pyplot as plt

# Plot daily average temperature in Celsius
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["TAVG_C"], color="blue")

plt.title("Daily Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")

# Save plot to file
plt.savefig("plots/daily_trend.png", dpi=300, bbox_inches="tight")

plt.show()
```
